```python
# Import Modules
import pandas as pd
```

```python
# Import CSV file
csv_file = "resources/realdonaldtrump.csv"
realdonaldtrump_df = pd.read_csv(csv_file)
```

```python
# Split date column into Day, Month, Year and Time
realdonaldtrump_df[['Date','Time']] = realdonaldtrump_df.date.str.split(" ",expand=True,)
realdonaldtrump_df[['Year','Month', 'Day']] = realdonaldtrump_df.Date.str.split("-",expand=True,)
realdonaldtrump_df.head()
```

```python
# Creating new trump tweets data frame
trump_tweets_df = realdonaldtrump_df[['id', 'link', 'content', 'retweets', 'favorites', 'mentions', 'hashtags', 'Date', 'Time', 'Year', 'Month', 'Day']]
trump_tweets_df.head()
```
