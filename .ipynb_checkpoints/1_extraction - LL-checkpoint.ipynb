Create new data by scraping the calendar website

```python
base_url = "https://www.timeanddate.com/holidays/us/"
```

```python
calendar = []
```

Scrape the calendar website and format the scraped data

```python
#Scrape the calendar for 2009 - 2020

for x in range (2009, 2021):
    url2 = (f'{base_url}{x}')
    tables = pd.read_html(url2)
    df = tables[0]
    df.columns = df.columns.droplevel(1)
    df["Year"] = x
    df = df.rename(columns = {"Unnamed: 1_level_0":"Day_of_Week"})
    df = df[:-1]
    calendar.append(df)

calendar
```

```python
# Convert the scrapped data to a dataframe

cb = pd.concat(calendar)
cb
```

```python
# Updated column heading

cb2 = cb["Date"].str.split(" ", n = 1, expand = True)
cb["Month"] = cb2[0]
cb["Day"] = cb2[1]
cb
```

```python
# Replace the month name with numbers

cb2 = cb.replace({"Jan":"01", "Feb":"02","Mar":"03","Apr":"04","May":"05","Jun":"06","Jul":"07","Aug":"08","Sep":"09","Oct":"10","Nov":"11","Dec":"12"})
cb2
```

```python
#Drop the "Date" column

cb3 = cb2.drop(columns = ["Date"])
cb3
```

```python
# Merge Year, Month, and Day column and convert to datetime format
cb3["Date"] = pd.to_datetime(cb3[["Year","Month","Day"]])
cb3
```
