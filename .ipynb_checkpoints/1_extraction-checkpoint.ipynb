```python
base_url = "https://www.timeanddate.com/holidays/us/"
```

```python
calendar = []
```

```python
for x in range (2009, 2021):
    url2 = (f'{base_url}{x}')
    tables = pd.read_html(url2)
    df = tables[0]
    df["Year"] = x
    calendar.append(df)

calendar
```
