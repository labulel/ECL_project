```python
base_url = "https://www.timeanddate.com/holidays/us/"
```

```python
calendar = []
```

```python
for x in range (2009, 2021):
    url2 = (f'{base_url}{x}')
    tables = pd.read_html(url2)
    df = tables[0]
    df.columns = df.columns.droplevel(1)
    df["Year"] = x
    df = df.rename(columns = {"Unnamed: 1_level_0":"Day"})
    df = df[:-1]
    calendar.append(df)

calendar
```

```python
df2 = calendar[2]
df2
```

```python
combined = pd.concat(calendar)
combined
```
